```python
from tensorflow.keras.preprocessing.text import one_hot  # type: ignore
```

```python
# Sentences
sent = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
sent
```

```python
# Define the vocabulary size
voc_size = 10000
```

```python
# One hot representation
one_hot_repr = [one_hot(word, voc_size) for word in sent]
one_hot_repr
```

```python
# Word Embedding Representation
from tensorflow.keras.layers import Embedding  # type: ignore
from tensorflow.keras.utils import pad_sequences  # type: ignore
from tensorflow.keras.models import Sequential  # type: ignore
import numpy as np
```

```python
sent_length = 8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
embedded_docs
```

```python
# Feature Representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size, dim, input_shape = (sent_length, )))
model.compile('adam', 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
