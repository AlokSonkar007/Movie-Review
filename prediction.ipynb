```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.datasets import imdb    # type: ignore
from tensorflow.keras.utils import pad_sequences  # type: ignore
from tensorflow.keras.models import load_model  # type: ignore
from tensorflow.keras.preprocessing import sequence  # type: ignore
import warnings
warnings.filterwarnings('ignore')
```

```python
# Reversing the index
word_index = imdb.get_word_index()
reverse_word_index = {value: key for key, value in word_index.items()}
reverse_word_index
```

```python
# Load the pre trained model
model = load_model('simple_rnn_imdb.h5')
model.summary()
```

```python
# All the weights used while training the model
model.get_weights()
```

```python
# Helper functions
# Function to decode reviews
def decode_reviews(encoded_review):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in encoded_review])

# Function to pre process user input
def preprocess_texts(text):
    words = text.lower().split()
    encoded_review = [word_index.get(word, 2) + 3 for word in words]
    padded_review = sequence.pad_sequences([encoded_review], maxlen=500)
    return padded_review
```

```python
# Prediction function
def predict_statement(review):
    preprocessed_input = preprocess_texts(review)

    prediction = model.predict(preprocessed_input)

    sentiment = 'Positive' if prediction[0][0] > 0.5 else "Negative"

    return prediction[0][0], sentiment
```

```python
# Step: 4 - User input and prediction
example_review = "The movie was fantastic ! The acting was great and the plot was thrilling"
sentiment, score = predict_statement(example_review)

print(f'Review: {example_review}')
print(f"Sentiment: {sentiment}")
print(f'Prediction Score: {score}')
```
